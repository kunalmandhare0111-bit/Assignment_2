```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 5)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, recall_score, f1_score

# -----------------------
# 0. Configuration
# -----------------------
# CORRECTED PATH: Use double backslashes for Windows paths
CSV_PATH = "C:\\Users\\Aasus\\Downloads\\DataCoSupplyChainDataset.csv"  # <-- CHANGE APPLIED
# -----------------------

# -----------------------
# 1. Load data
# -----------------------
# FIX: Added encoding='latin1' to resolve UnicodeDecodeError
df = pd.read_csv(CSV_PATH, low_memory=False, encoding='latin1') 
print("Loaded data shape:", df.shape)

# -----------------------
# 2. Create targets
# -----------------------
# Create binary target for fraud and late delivery as in the assignment hint
df["fraud"] = np.where(df.get("Order Status") == "SUSPECTED_FRAUD", 1, 0)
df["late_delivery"] = np.where(df.get("Delivery Status") == "Late delivery", 1, 0)

# -----------------------
# 3. Cleaning & feature preparation
# -----------------------
# Work on a deep copy to avoid chained assignment issues
df_clean = df.copy(deep=True)

# Drop leakage / identifiers (adjust if your dataset has different column names)
cols_to_drop = [
    "Order Status", "Delivery Status",
    "Order Id", "Customer Id", "Customer Email",
    "Product Name", "Order Item Id"
]
cols_to_drop = [c for c in cols_to_drop if c in df_clean.columns]
df_clean = df_clean.drop(columns=cols_to_drop)

# Drop columns that are completely NaN
all_null_cols = df_clean.columns[df_clean.isna().all()]
if len(all_null_cols) > 0:
    print("Dropping all-NaN columns:", list(all_null_cols))
    df_clean = df_clean.drop(columns=all_null_cols)

# Fill remaining NaNs safely (categorical -> mode, numeric -> median)
for col in df_clean.columns:
    if df_clean[col].dtype == "O":
        # if column all NaN, fallback handled above
        df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
    else:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())

print("Total remaining NaNs:", df_clean.isna().sum().sum())

# Label-encode categorical columns
cat_cols = df_clean.select_dtypes(include=["object"]).columns.tolist()
for col in cat_cols:
    le = LabelEncoder()
    df_clean[col] = le.fit_transform(df_clean[col])

# -----------------------
# 4. Build features and scale
# -----------------------
target_cols = ["fraud", "late_delivery"]
feature_cols = [c for c in df_clean.columns if c not in target_cols]

X = df_clean[feature_cols].values
y_fraud = df_clean["fraud"].values
y_late = df_clean["late_delivery"].values

# Remove constant (zero-variance) features
var_sel = VarianceThreshold(threshold=0.0)
X = var_sel.fit_transform(X)
feature_cols = np.array(feature_cols)[var_sel.get_support()]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Feature matrix shape after scaling:", X_scaled.shape)


# -----------------------
# 5. Models & evaluation helpers
# -----------------------
def create_models():
    """Return a fresh dict of models."""
    return {
        "Logistic regression": LogisticRegression(max_iter=1000),
        "L1 Logistic (Lasso)": LogisticRegression(penalty="l1", solver="liblinear", max_iter=1000),
        "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1),
        "Gradient Boosting": GradientBoostingClassifier(random_state=42)
    }


def evaluate_models(X_train, X_test, y_train, y_test):
    """
    Train the standard set of models on X_train/y_train and evaluate on X_test/y_test.
    Returns dict: model_name -> (accuracy, recall, f1).
    """
    models = create_models()
    results = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        rec = recall_score(y_test, y_pred, pos_label=1)
        f1 = f1_score(y_test, y_pred, pos_label=1)
        results[name] = (acc, rec, f1)
    return results


# -----------------------
# 6. Compare 80-20 vs 70-30 splits and keep the best
# -----------------------
splits = {"80-20": 0.2, "70-30": 0.3}
all_results = {}

for split_name, test_size in splits.items():
    # FRAUD split
    X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(
        X_scaled, y_fraud, test_size=test_size, random_state=42, stratify=y_fraud
    )
    fraud_res = evaluate_models(X_train_f, X_test_f, y_train_f, y_test_f)

    # LATE DELIVERY split
    X_train_l, X_test_l, y_train_l, y_test_l = train_test_split(
        X_scaled, y_late, test_size=test_size, random_state=42, stratify=y_late
    )
    late_res = evaluate_models(X_train_l, X_test_l, y_train_l, y_test_l)

    # Build result rows
    rows = []
    model_names = list(create_models().keys())
    for name in model_names:
        acc_f, rec_f, f1_f = fraud_res[name]
        acc_l, rec_l, f1_l = late_res[name]
        rows.append([name, acc_f, rec_f, f1_f, acc_l, rec_l, f1_l])

    cols = [
        "Classification Model",
        "Accuracy - Fraud", "Recall - Fraud", "F1 - Fraud",
        "Accuracy - Late", "Recall - Late", "F1 - Late"
    ]
    results_df = pd.DataFrame(rows, columns=cols)

    all_results[split_name] = {
        "df": results_df,
        "fraud": (X_train_f, X_test_f, y_train_f, y_test_f),
        "late": (X_train_l, X_test_l, y_train_l, y_test_l)
    }

# Choose best split based on mean F1 across both targets
best_split = None
best_score = -np.inf
for split_name, obj in all_results.items():
    df_res = obj["df"]
    mean_f1 = (df_res["F1 - Fraud"].mean() + df_res["F1 - Late"].mean()) / 2.0
    if mean_f1 > best_score:
        best_score = mean_f1
        best_split = split_name

best_results_df = all_results[best_split]["df"]

print("\n=========================")
print(f"Best split selected: {best_split} (mean F1 = {best_score:.4f})")
print("=========================\n")
print(best_results_df.to_string(index=False))

# Save Table-4 style results for the best split to CSV for your report
best_results_df.to_csv("table4_best_split.csv", index=False)
print("Saved best split results to table4_best_split.csv")

# -----------------------
# 7. Train best model for each target (from best split)
# -----------------------
X_train_f, X_test_f, y_train_f, y_test_f = all_results[best_split]["fraud"]
X_train_l, X_test_l, y_train_l, y_test_l = all_results[best_split]["late"]

fraud_res = evaluate_models(X_train_f, X_test_f, y_train_f, y_test_f)
late_res = evaluate_models(X_train_l, X_test_l, y_train_l, y_test_l)

best_model_fraud_name = max(fraud_res, key=lambda k: fraud_res[k][2])  # F1 index = 2
best_model_late_name = max(late_res, key=lambda k: late_res[k][2])

print("\nBest model for FRAUD:", best_model_fraud_name, "metrics:", fraud_res[best_model_fraud_name])
print("Best model for LATE DELIVERY:", best_model_late_name, "metrics:", late_res[best_model_late_name])

# instantiate and fit the chosen best models
models_map = create_models()
best_model_fraud = models_map[best_model_fraud_name]
best_model_late = models_map[best_model_late_name]

best_model_fraud.fit(X_train_f, y_train_f)
best_model_late.fit(X_train_l, y_train_l)


# -----------------------
# 8. Feature importance (top 5) and plots
# -----------------------
def get_feature_importance(model, feature_names):
    if hasattr(model, "feature_importances_"):
        importances = model.feature_importances_
    elif hasattr(model, "coef_"):
        importances = np.abs(model.coef_).ravel()
    else:
        raise ValueError("Model does not provide feature importance.")
    return pd.Series(importances, index=feature_names).sort_values(ascending=False)


# FRAUD feature importances
fi_fraud = get_feature_importance(best_model_fraud, feature_cols)
top5_fraud = fi_fraud.head(5)
print("\nTop 5 features for FRAUD detection:")
print(top5_fraud)

plt.figure()
top5_fraud.sort_values().plot(kind="barh")
plt.title(f"Top 5 Feature Importances - Fraud ({best_model_fraud_name})")
plt.xlabel("Importance")
plt.tight_layout()
plt.savefig("feature_importance_fraud.png")
plt.show()

# LATE DELIVERY feature importances
fi_late = get_feature_importance(best_model_late, feature_cols)
top5_late = fi_late.head(5)
print("\nTop 5 features for LATE DELIVERY prediction:")
print(top5_late)

plt.figure()
top5_late.sort_values().plot(kind="barh")
plt.title(f"Top 5 Feature Importances - Late Delivery ({best_model_late_name})")
plt.xlabel("Importance")
plt.tight_layout()
plt.savefig("feature_importance_late.png")
plt.show()

# Save top-5 lists to CSV for convenience
top5_fraud.to_csv("top5_features_fraud.csv", header=["importance"])
top5_late.to_csv("top5_features_late.csv", header=["importance"])
print("\nSaved feature importance plots and CSVs.")
```
